```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for age distribution
np.random.seed(0)  # for reproducibility
ages = np.random.normal(loc=30, scale=10, size=1000)  # Example ages

# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=30, color='salmon', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for gender distribution
genders = ['Male', 'Female', 'Non-binary', 'Prefer not to say']
counts = [460, 510, 30, 20]

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(genders, counts, color='skyblue')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
